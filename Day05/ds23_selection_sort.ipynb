### 선택 정렬

#### 최소값 위치 확인 함수 

```python
def findMinIdx(ary):
    minIdx = 0                      # 배열의 인덱스는 0부터 시작
    for i in range(1, len(ary)):
        if (ary[minIdx])>ary[i]:    # 지정된 값이 배열에 현재 값보다 크면  안됌
            minIdx = i 
    
    return minIdx
```

```python
tests = [55, 88, 33, 77, 10]
print(findMinIdx(tests))
```

## 선택 정렬 구현 

```python
# 변수 선언
before = [188, 162, 168, 120, 50, 150, 177, 105]
after = []

print('정렬 전 --->', before)

for _ in range(len(before)):
    minPos = findMinIdx(before)
    after.append(before[minPos])
    del(before[minPos])          # 최소값 삭제

print('정렬 후 -->', after)
```

#### 개선된 선택 정렬

배열을 두개 만들 필요 없고 값 교환하는 swap 기능을 학습

```python
def selectionSort(ary):
    n = len(ary)
    for i in range(0, n-1):            # 마지막 인덱스가 n-2
        minIdx = i
        for k in range(i+1, n):  
            if (ary[minIdx]>ary[k]):   # 마지막 인덱스 n-1
                minIdx = k
        # tmp = ary[i]
        # ary[i] = ary[minIdx]          # 최소값으로 변경
        # ary[minIdx]=tmp               # 구식  
        ary[i], ary[minIdx]=ary[minIdx], ary[i]
    
    return ary

# 변수
dataAry=[188, 162, 168, 120, 50, 150, 177, 105]

print('정렬 전 : ', dataAry)
dataAry = selectionSort(dataBank)
print('정렬 후 : ',  dataBank)

# 1000 5점 4초
```

### 삽입정렬  Insertion Sort

#### 자신의 삽입 위치를 찾는 함수 

```python
def findInsertIdx(ary, data):
    findIdx= -1
    for i in range(0, len(ary)):
        if ary[i] >  data :
            findIdx = i 
            break
    if findIdx == -1:
        return len(ary)
    else:
        return findIdx
```

```python
testAry=[]
print('50이 들어갈 위치 ',findInsertIdx(testAry, 50))
testAry = [33, 60, 50, 77, 88]
print('55가 들어갈 위치', findInsertIdx(testAry, 55))
```

### 삽입 정렬 개선

```python
def insertionSort(ary):
    n =len(ary)
    for end in range(1, n):             # 앞의 값 하나 빼고 끝까지 반복
        for cur in range(end, 0, -1):   # 반대로 돌면서
            if ary[cur-1] > ary[cur]:   # 오름차순
                ary[cur-1], ary[cur]=ary[cur], ary[cur-1]    # 파이썬 swap 
    return ary

dataAry=[188, 162, 168, 120, 50, 150, 177, 105]

print('정렬 전 : ', dataAry)
dataAry = insertionSort(dataAry)
print('정렬 후 : ',  dataAry)
```

### 버블 정렬 (bubble sort)

```python
def BubbleSort(ary):
    n = len(ary)
    for end in range(n-1, 0, -1):    # 반대로 반복
        print('싸이클# -->', ary)
        for cur in range(0, end):    # 0부터 n-2까지 반복 
            if ary[cur] >ary[cur+1]:
                ary[cur], ary[cur+1]= ary[cur+1], ary[cur]
    return ary

dataAry=[190, 162, 168, 120, 50, 150, 177, 105]

print('정렬 전 : ', dataAry)
dataAry = BubbleSort(dataAry)
print('정렬 후 : ',  dataAry)
```

퀵 정렬(Quick Sort)

```python
def quickSort(ary):
    n = len(ary)
    if n  <= 1:           # 정렬 개수가 1개 이하면
        return ary
    pivot = ary[n//2]     # 중간값을 기준값으로 
    leftAry, rightAry, midAry = [], [], []
    
    for num in ary :
        if num < pivot:
            leftAry.append(num)
        elif num > pivot:
            rightAry.append(num)
        else:
            midAry.append(num)     # 중복을 고려 
        
    return quickSort(leftAry) + midAry + quickSort(rightAry)

dataAry=[190, 162, 168, 120, 50, 150, 177, 105]

print('정렬 전 : ', dataAry)
dataAry = quickSort(dataAry)
print('정렬 후 : ',  dataAry)
```
